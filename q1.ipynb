```
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

First of all , we read the image as a grayscale image and setting the filter as a parameter in order to use it.

Then, we apply the filter and show the image after the application.

```
fig, (before_plot, after_filter_plot) = plt.subplots(1, 2)

before_plot.set_title('before filter')
after_filter_plot.set_title('after filter')

path = 'petrucci.jpg'
img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
before_plot.imshow(img, cmap='gray')

filter = np.array([[1, 1, 1],
                   [1, -8, 1],
                   [1, 1, 1]])

filtered_image = cv2.filter2D(img, -1, filter)
after_filter_plot.imshow(filtered_image, cmap='gray')

plt.show()
```
