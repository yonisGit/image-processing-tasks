First of all, let's import all the needed libraries:

```
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

**Note: to show the image I used matplotlib that shows the images on the order rgb, and the openCV library is using the bgr order, so before showing the images I performed the needed conversions using the openCV library.**

Reading the file with opencv, 
then converting the image from bgr to rgb (in opencv the order of the colors is bgr and on matplotlib its rgb)
and then showing the image of the best guitarist in the world:

```
path = 'petrucci.jpg'
image = cv2.imread(path)
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
```

Then, we are going to split the image to colors rgb.

**Notice: in openCV the order of the colors is bgr.**

```
b, g, r = cv2.split(image)
```

Now it's time to merge the colors by the needed order (swap red with the green).

```
new_image=cv2.merge((b,r,g))
```

Lets show the new image after the swap:

```
new_image=cv2.cvtColor(new_image,cv2.COLOR_BGR2RGB)
plt.imshow(new_image)
```

And now lets save it.

```
plt.imsave('OutPicture.png',new_image)
```
